# Titanic DataSet
You have been provided with the Titanic Data Set. The dataset shows the survival rate of passsengers in the Titanic crash which is the real dataset. The Dataset is downloaded from kaggle. In the dataset provided, you may ignore the columns in the red font (i.e. A,D,G,H,I,K,L) and use the remaining columns to prdict the survival rate.

The "Survived" column is the target variable. You have to predict whetehr the passenger survived or not based on the: Class (Column 'C'), Sex (Column 'E'), Age (Column 'F') and Fare the passenger paid before onboarding the Titanic ship(Column 'J'). Come up witha  score.

```python
import pandas as pd
df = pd.read_csv("titanic.csv")
df.head()
```

```python
inputs = df.drop(['PassengerId', 'Name', 'SibSp', 'Parch', 'Ticket', 'Cabin', 'Embarked'], axis = 'columns')
inputs = inputs.dropna(axis=0)
target = inputs['Survived']
inputs = inputs.drop('Survived', axis='columns')
inputs
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_sex = LabelEncoder()
```

```python
inputs['sex_n'] = le_sex.fit_transform(inputs['Sex'])
inputs.head()
```

```python
inputs_n = inputs.drop(['Sex'], axis = 'columns')
inputs_n
```

```python
target
```

```python
from sklearn import tree
```

```python
model = tree.DecisionTreeClassifier()
```

```python
model.fit(inputs_n, target)
```

```python
model.score(inputs_n,target)
```

```python
model.predict([[2, 59.0, 27, 0]])
```
